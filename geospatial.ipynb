```python
import geopandas as gpd
import pandas as pd
```

```python

# Load the world map data from GeoPandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Save the GeoJSON file
output_geojson_path = './data/world_countries.geojson'
world.to_file(output_geojson_path, driver='GeoJSON')
```

```python
def findMissing(data):
    print(f'Total Rows: {data.shape[0]}\n')
    for c in data.columns:
        missing = data.shape[0] - data[c].describe().loc['count']
        if missing>0:
            print('-----------------------')
            print(c)
            print('-----------------------')
            print(data[c].describe())
            print(f'\nMissing Values: {int(missing)}\n')
```

```python
df = pd.read_csv('./data/clean_data.csv')
world = gpd.read_file('./data/world_countries.geojson')
```

```python
# Merge the CSV data with the GeoJSON data based on country names
merged_data = world.merge(df, how='inner', left_on='name', right_on='country')

# Save the merged GeoJSON file
output_geojson_path = './data/world.geojson'
merged_data.to_file(output_geojson_path, driver='GeoJSON')
```
